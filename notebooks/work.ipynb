```python
# 1. Load the CSV correctly using pd.read_csv
df = pd.read_csv(r"C:\Users\PC\Documents\projecto_py\projecto_py\data\lab_limpo.csv")

# 2. Create the pivot table (notice the quoted column names and 'aggfunc')
tabela_dinamica = df.pivot_table(
    values='id_amostra', 
    index='resultado', 
    columns='id_paciente', 
    aggfunc='sum', 
    fill_value=0
)
print (tabela_dinamica)
```

```python
x = df["valor_ct"]
media = np.mean(x)
dp = np.std(x, ddof=1)
n = len(x)
ic_inf = media - 1.96 * (dp / np.sqrt(n))
ic_sup = media + 1.96 * (dp / np.sqrt(n))
ic_inf, ic_sup
```

```python
plt.plot(df["resultado"], df["qualidade_amostra"], "o") 
plt.title("Relação entre Resultado e Qualidade da Amostra")
plt.tight_layout()
plt.show()
```

```python
plt.title("Distribuição dos valores Ct")
plt.xlabel("valores_ct")
plt.ylabel("Frequência")
df.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Histograma
plt.hist(
    df["valor_ct"],
    bins=15,
    color="steelblue",
    edgecolor="black",
    alpha=0.8
)

# Personalização
plt.title(
    "Distribuição dos Valores Ct",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Valor Ct",
    fontsize=12
)
```

```python
sns.boxplot(
    data=df,
    x="resultado",
    y="valor_ct"
)

plt.show()
```

```python
sns.boxplot(
    data=df,
    x="laboratorio",
    y="valor_ct"
)

plt.show()
```

```python
tabela = (
    df.groupby("patogeno")
      .agg(
          n=("valor_ct","count"),
          media_ct=("valor_ct","mean"),
          dp_ct=("valor_ct","std"),
          mediana_ct=("valor_ct","median")
      )
)

print(tabela.round(2))
```

```python
plt.figure(figsize=(8,6))

sns.violinplot(
    data=df,
    x="patogeno",
    y="valor_ct"
)

plt.ylabel("Ct")
plt.xlabel("Patógeno")

plt.savefig(
    "figura_ct_patogeno.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
